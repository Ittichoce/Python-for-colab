# **Working Different data types in same environment**

```python
print(type(78+22))
print(type(78+22 * 5))
print(type(12.5 - 1.0))
print(type(2+3j + 1 - 2j))
```

We can see that binary Arithmetic operators with *both operands being of similar data types* result in values with the *same data types of the operands.*

Except for the *division operator*.

```python
print(type(2 / 2))
print(type(2.0 / 2.0))
```

We can see that whether the operands were thier either `int` or `float`, the result were always `float`. 

This makes sense since the result of a division can have *decimal points*.

Now, let's try dividing complex numbers.

```python
c1 = 4+4j
c2 = 2+2j
print(c1 / c2)
print(type(c1 / c2))
```

This again, made sense since a division of two complex numbers can result in another complex number.

##**Implicit Data type Conversion**

image.png

Implicit Data Type Conversion
- Occurs *automatically* 
- **Implicit Data Type Conversion** alaways promote data type capable of holding *less information* to data type capable of holding *more information.*

```python
print(type(32 * 5.0 ))
```

In the above statement, the `int` value of 32 was **converted to** `float` and then **mutiplied** with 5.0.

```python
print(type(5 * (3+2j)))
print(type((1+1j) + -0.5))
```

In the code above, both `int` and `float` were converted to `complex` before being used as operands in their respective statements.

### **Explicit Data Type Conversoion** 
We use *typecasting functions* to convert data to the desired data types.
Typecasting  function include (but not limited to):
- `int()`
- `float()`
- `complex()`
- `str()`
- `bool()` 

**An example situation in which *explicit data type conversion* is needed.**

```python
#This is program claculates n^10, where n is a number entered by the user. 
n = input("Please enter n: ")    # n is string.
result = float(n) ** 10  
#Show the output as n = [the input] --> n^10 = [the result]
print("n = " + n + " --> n^10 = " + str(result)) #change type of result from float to string. while n is string.
```
